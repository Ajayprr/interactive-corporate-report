```python
import spacy
import numpy as np

nlp = spacy.load('en')

sample_review = ""
for i in df_top_react["status_message"]:
    print(str(i)[2:-3].rsplit(' ', 1)[0])
    sample_review = sample_review + " " + str(i)[2:-3].rsplit(' ', 1)[0]

# print(sample_review)

len(sample_review)

sample_review = sample_review.replace("\\", "")
sample_review = sample_review.replace("â","'")
sample_review = sample_review.replace("ð","")
sample_review = sample_review.replace("ââï","")
sample_review = sample_review.replace("ðâï","")
sample_review = sample_review.replace("â","")
sample_review = sample_review.replace('¬†', '')
sample_review = sample_review.replace('.', '')
sample_review = sample_review.replace('‚Äô','')
sample_review = sample_review.replace('‚Ä†','')
sample_review = sample_review.replace('√±','n')
sample_review = sample_review.replace('√©','i')
sample_review = sample_review.replace('Ôøº','')
sample_review = sample_review.replace('Ñ¢','')
sample_review = sample_review.replace('¬Æ','')

# sample_review = sample_review.replace("xe2x80xa6"," ")

parsed_review = nlp(sample_review)

token_text = [token.orth_ for token in parsed_review]
token_pos = [token.pos_ for token in parsed_review]

df = pd.DataFrame({'token_text': token_text, 'part_of_speech': token_pos})
```

```python
list_dates = [] 
for coy in codes:
    fb_frame = pd.read_csv(path_out+coy + "_facebook.csv",encoding='unicode-escape')

    fb_frame['status_published'] = fb_frame['status_published'].apply(lambda x: parse(x))

    list_dates.append(fb_frame["status_published"].min().year)

max(list_dates)+1

coy = "BJRI"
rap = pd.DataFrame(columns=["React (R)","Comm (C)","Share (S)","S/R","C/R","S/C"],index=codes)
rap_q = pd.DataFrame(columns=["React (R)","Comm (C)","Share (S)","S/R","C/R","S/C"],index=codes)
rap_q2 = pd.DataFrame(columns=["React (R)","Comm (C)","Share (S)","S/R","C/R","S/C"],index=codes)




for coy in codes:

    fb_frame = pd.read_csv(path_out+coy + "_facebook.csv",encoding='unicode-escape')

    fb_frame['status_published'] = fb_frame['status_published'].apply(lambda x: parse(x))


    date_cut = str(max(list_dates)+1)+"-01-01"

    date_cut

    qtr_1_start = datetime.now() - timedelta(90)
    qtr_2_start = qtr_1_start - timedelta(90)

    frame_q1 = fb_frame[fb_frame["status_published"]>qtr_1_start].reset_index(drop=True)

    frame_q2 = fb_frame[(fb_frame["status_published"]>qtr_2_start) & (fb_frame["status_published"]<qtr_1_start)].reset_index(drop=True)

    frame_all = fb_frame[fb_frame["status_published"]>date_cut].reset_index(drop=True)

    frame_q1["status_type"].value_counts()

    num_reactions=frame_all.sum()["num_reactions"]
    num_comments=frame_all.sum()["num_comments"]
    num_shares=frame_all.sum()["num_shares"]
    
    num_reactions_q=frame_q1.sum()["num_reactions"]
    num_comments_q=frame_q1.sum()["num_comments"]
    num_shares_q=frame_q1.sum()["num_shares"]
    
    num_reactions_q2=frame_q2.sum()["num_reactions"]
    num_comments_q2=frame_q2.sum()["num_comments"]
    num_shares_q2=frame_q2.sum()["num_shares"]
    
    num_reactions_v=frame_q1[frame_q1["status_type"]=="video"].sum()["num_reactions"]
    num_comments_v=frame_q1[frame_q1["status_type"]=="video"].sum()["num_comments"]
    num_shares_v=frame_q1[frame_q1["status_type"]=="video"].sum()["num_shares"]

    num_reactions_p=frame_q1[frame_q1["status_type"]=="photo"].sum()["num_reactions"]
    num_comments_p=frame_q1[frame_q1["status_type"]=="photo"]["num_comments"]
    num_shares_p=frame_q1[frame_q1["status_type"]=="photo"].sum()["num_shares"]

    s_r = num_shares/num_reactions
    c_r = num_comments/num_reactions
    s_c = num_shares/num_comments
    
    s_r_q = num_shares_q/num_reactions_q
    c_r_q = num_comments_q/num_reactions_q
    s_c_q = num_shares_q/num_comments_q
    
    s_r_q2 = num_shares_q2/num_reactions_q2
    c_r_q2 = num_comments_q2/num_reactions_q2
    s_c_q2 = num_shares_q2/num_comments_q2

    s_r_v = num_shares_v/num_reactions_v
    c_r_v = num_comments_v/num_reactions_v
    s_c_v = num_shares_v/num_comments_v

    s_r_p = num_shares_p/num_reactions_p
    c_r_p = num_comments_p/num_reactions_p
    s_c_p = num_shares_p/num_comments_p

    rap.loc[coy,:] = [num_reactions,num_comments,num_shares,round(s_r,2),round(c_r,2),round(s_c,2)]
    rap_q.loc[coy,:] = [num_reactions_q,num_comments_q,num_shares_q,round(s_r_q,2),round(c_r_q,2),round(s_c_q,2)]
    rap_q2.loc[coy,:] = [num_reactions_q2,num_comments_q2,num_shares_q2,round(s_r_q2,2),round(c_r_q2,2),round(s_c_q2,2)]
```

```python
rl =  pd.concat((rap, rap_q2),axis=1)

rl = pd.concat((rl, rap_q),axis=1)
```
